```python
import numpy as np
import pandas as pd

import requests as rq
from bs4 import BeautifulSoup as bs
from scrapy import Selector
from selenium import webdriver

#preprocessing
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

#nltk
import nltk
nltk.download('punkt_tab')
nltk.download("wordnet")
nltk.download("stopwords")
#from nltk.stem import * # Removed this line to prevent potential conflicts
from nltk.corpus import stopwords

from sklearn.metrics import pairwise_distances
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
#TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
#KNN
from sklearn.neighbors import KNeighborsClassifier
```

# Preprocessing

```python
stemmer=StemmerFactory().create_stemmer()
stopword=StopWordRemoverFactory().create_stop_word_remover()

#use request with header (user-agent)
headers={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

#if search query: "kecerdasan buatan"
url = 'https://scholar.google.co.id/scholar?hl=id&as_sdt=0%2C5&q=kecerdasan+buatan&btnG='

try:
  page = rq.get(url, headers=headers)
  print(page.status_code)

  if page.status_code == 200:
    extract = bs(page.content, 'html.parser') #ubah html jadi struktur yg bisa dicari

    results=extract.find_all("div", {"class":"gs_r gs_or gs_scl"})
    i=0
    for item in results:
      if i==10:break

      title=item.find("h3", {"class":"gs_rt"})
      if title:
        raw_title=title.get_text()

        a=title.find("a")
        link=a.get("href") if a else "no link"

        stemJudul=stemmer.stem(raw_title)
        stopJudul=stopword.remove(stemJudul)
        print(f"Original: {raw_title}")
        print(f"Stemmed : {stemJudul}")
        print(f"StopRem : {stopJudul}")
        print(f"Link    : {link}")
        print("-" * 30)

        i += 1
  else:
    print("Failed to retrieve page")
except Exception as e:
  print(f"Error: {e}")
```
